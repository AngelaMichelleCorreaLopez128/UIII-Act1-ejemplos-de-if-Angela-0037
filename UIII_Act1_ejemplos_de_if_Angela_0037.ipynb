#### Ejemplo de if

```python
# Ejemplo if
puntos_vida = 10  # Asigna el valor 10 a la salud del personaje
if puntos_vida < 20:  # Evalúa si la vida es menor a 20 usando el operador <
    print("¡Cuidado! Tu salud es baja.")  # Se ejecuta solo si la condición es verdadera
```

```python
# Ejemplo if-else
numero = 7

if numero % 2 == 0:
    print("El número es par.")
else:
    print("El número es impar.")
```

```python
# Ejemplo if-else
numero = 6

if numero % 2 == 0:
    print("El número es par.")
else:
    print("El número es impar.")
```

```python
# Ejemplo if-elif
edad = 8

if edad < 13:
    print("Categoría: Niño")
elif edad < 18:
    print("Categoría: Adolescente")
else:
    print("Categoría: Adulto")
```

```python
# Ejemplo if-elif
edad = 14

if edad < 13:
    print("Categoría: Niño")
elif edad < 18:
    print("Categoría: Adolescente")
else:
    print("Categoría: Adulto")
```

```python
# Ejemplo if-elif
edad = 22

if edad < 13:
    print("Categoría: Niño")
elif edad < 18:
    print("Categoría: Adolescente")
else:
    print("Categoría: Adulto")
```

```python
# Ejemplo if-elif
# Ejemplo: Categorizar la edad usando múltiples comparaciones
edad = int(input("Ingrese su edad: "))
if edad < 13:
    print("Usted es un niño.")
elif edad < 18:
    print("Usted es un adolescente.")
else:
    print("Usted es un adulto.")
```

```python
año = 2026  # Define la variable anio con el valor 2026
if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):  # Combina operadores aritméticos, relacionales y lógicos (and, or)
    print(f"{año} es bisiesto.")  # Imprime el resultado si el año cumple las reglas de bisiesto
else:  # Bloque para años que no cumplen los requisitos lógicos establecidos
    print(f"{año} no es bisiesto.")  # Informa que el año ingresado es un año común
```
